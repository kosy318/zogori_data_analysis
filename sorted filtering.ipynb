```python
import os
dir = './sorted/'
```

```python
folder_list = os.listdir(dir)
folder_list
```

```python
file_list = {}
for folder in folder_list[1:]:
    file_list[folder] = os.listdir(f'{dir}{folder}')
```

```python
file_list
```

```python
for folder in folder_list[1:]:
    bf_num = '00'
    bf = ''
    for file in os.listdir(f'{dir}{folder}'):
        if file.split('.')[-1] != 'cpp' and file.split('.')[-1] != 'c':
            print('not c/c++: delete', file)
            os.remove(f'{dir}{folder}/{file}')
            continue
            
        if file.split('_')[1] != folder:
            print('problem incorrect: delete', file)
            os.remove(f'{dir}{folder}/{file}')
            continue
            
        if file.split('.')[-2][-1] != 'c' and file.split('.')[-2][-1] != 'C':
            print('delete', file)
            os.remove(f'{dir}{folder}/{file}')
            continue
            
        if bf_num == file.split('.')[-2][:2]:
            print('delete', bf, 'leave', file)
            os.remove(f'{dir}{folder}/{bf}')
        
        bf_num = file.split('.')[-2][:2]
        bf = file
```
